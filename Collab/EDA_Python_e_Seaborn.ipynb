[Link](https://www.kaggle.com/datasets/abcsds/pokemon)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('/content/Pokemon.csv')
```

```python
df
```

```python
df.info()
```

```python
#Retirar primeira coluna
del df['#']
```

```python
#Próximo passo é entender se o dataset apresenta valores nulos - por padrão do data set 
df.isna().sum()
```

```python
#Tratando nulos
df['Type 2'] = df['Type 2'].fillna('null')
```

```python
df = df.rename(columns={"Name":"name",
                   "Type 1":"type_1", 
                   "Type 2":"type_2", 
                   "Total":"total", 
                   "HP":"hp", 
                   "Attack":"attack",     
                   "Defense":"defense", 
                   "Sp. Atk":"sp_atk", 
                   "Sp. Def":"sp_def",
                   "Speed":"speed", 
                   "Generation":"generation",
                   "Legendary":"legendary"})

df
```

```python
round(df.describe())
```

```python
df_cor = df.corr()
```

```python
plt.figure(figsize=(15, 5))
plt.title('Heatmap')
sns.heatmap(df_cor, annot = True, vmin= -1, vmax=1, cmap = 'coolwarm')
```

```python
#Quantidade de Pokemons por tipo1
plt.figure(figsize=(15, 5))

plt.title('Qtd. de Pokémons por Tipo 1')

sns.countplot(data = df, 
              x='type_1', 
              order = df['type_1'].value_counts().index)
```

```python
##Quantidade de Pokemons por tipo2
plt.figure(figsize=(15, 5))
df_type2 = df[df['type_2'] != 'null']

plt.title('Qtd. de Pokémons por Tipo 2')
sns.countplot(data = df_type2, 
              x='type_2', 
              order = df['type_2'].value_counts().index)
```

```python
##Quantidade de Pokemon por Geração
plt.figure(figsize=(15, 5))

plt.title('Qtd. de Pokémons por Geração')
sns.countplot(data = df, 
              x='generation', 
              order = df['generation'].value_counts().index)
```

```python
##Quantidade de Pokemon Legendary
plt.figure(figsize=(10, 5))

plt.title('Qtd. de Pokémons Lendário')

sns.countplot(data = df, 
              x='legendary', 
              order = df['legendary'].value_counts().index)
```

```python
#percentual
df_perc = df['legendary'].value_counts(normalize = True)
round(df_perc*100)
```

```python
df
```

```python
#Velocidade
sns.displot(x = 'speed',
            data = df,
            kde = True,
          ).set(title = "Distribuição de Velocidade")
```

```python
#Velocidade por geração
plt.figure(figsize = (15,10))
sns.displot(x = 'speed',
            data = df,
            kde = True,
            col = 'generation'
          )
```

```python
df
```

```python
#Dual type

#dual type e 
na = df[df["type_2"] == 'null']['name'].count()
not_null = df[df["type_2"] != 'null']['name'].count()

total = (na + not_null)

na = round((na/total)*100)
notnull = round((not_null/total)*100)

dual = [na, notnull]

#Convertendo para DF
df_dual = pd.DataFrame(dual)

#Setando o Index
df_dual.index = ['na', 'not_null']

#Setando o Nome da Coluna
df_dual.rename(columns={0:"total"}, inplace = True)

#Criando o gráfico
plt.style.use('ggplot')
plt.figure(figsize = (15,5))
plot = df_dual.plot(kind = 'bar', title = '% Pokémons Dual Type', color = 'orange')

for i in plot.patches:
  plot.annotate(i.get_height(),
               (i.get_x() + i.get_width()/ 2, i.get_height()),
                ha = 'center')
```

```python
##Pokemóns lendários tem o atack e a defesa mais forte?

graf = sns.relplot(data = df,
            x = 'sp_atk',
            y = 'sp_def',
            hue = 'legendary',
            height = 6.5,
            palette = ['orange','blue']
            ).set(title = 'Dispersão')
```

```python
#lendários por força total
sns.displot(x = 'total',
            data = df,
            kind ='kde',
            height = 7,
            hue = 'legendary'
          ).set(title = "Distribuição de Força Total")
```
